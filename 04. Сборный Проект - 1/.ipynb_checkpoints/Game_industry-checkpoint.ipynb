### Часть 2. Предобработка данных  <a id='2.'></a>

Исправим ошибки в данных и преобразуем данные к нужным типам.

#### 2.1. Приведение названий столбцов к единому регистру <a id='2.1.'></a>

Рассмотрим названия всех столбцов датасета и преобразуем их к единому регистру.

```python
# Выведем на экран названия столбцов датафрейма
df.columns
```

```python
# Приведем все названия столбцов к нижнему регистру
df.columns = df.columns.str.lower() 
df.head()
```

```python
# Приведем названия консолей к верхнему регистру
df['platform'] = df['platform'].str.upper()
df.head()
```

```python
# Приведем жанры ирг к нижнему регистру
df['genre'] = df['genre'].str.lower()
df.head()
```

#### 2.2. Преобразование данных к нужным типам <a id='2.2.'></a>

Приведение столбца year_of_release, отражающего год выпуска игры, к целочисленном типу не только улучшит визуальное восприятие информации, но и сэкономит память и время при дальнейших расчетах.

```python
# Приведем данные столбца year_of_release к целому числу
df['year_of_release'] = df['year_of_release'].astype('Int64')
df['year_of_release'].dtypes
```

По аналогии с оценкой критиков, приведем оценку пользователей или user_score к числовому значению с дробной частью, что позволит максимально корректно отображать коэффициент каждой игры.

```python
# Посмотрим все уникальные значения столбца user_score
df['user_score'].unique()
```

```python
# Изучим отдельно строки датафрейма где оценка пользователей принимает значение "tbd"
tbd = df.query('user_score == "tbd"')
tbd.sample(5)
```

```python
# Заменим значение "tbd" на NaN
df['user_score'] = df['user_score'].replace({'tbd': np.nan})
df.sample(5)
```

```python
# Приведем данные столбца user_score к дробному числу
df['user_score'] = df['user_score'].astype('float64')
df['user_score'].dtypes
```

Одним из уникальных значений в оценке пользователей было обнаружено - tbd, что означает "To Be Determined" или "Не определено". Как правило, такое значение присваивается, если игра имеет минимальное количество отзывов, которое пока не позволяет дать объективную оценку.  
Руководствуясь этой деталью, заменили tbd на аналогичный по смыслу NaN,  что в дальнейшем позволило привести данные из строковых к нужному нам типу - дробному числу.

#### 2.3. Обработка пропусков в данных <a id='2.3.'></a>

Количество пропусков в столбцах name и genre совпадает, что может говорить об их отсутствии в одинаковых строках датасета, и настолько мало, что от них можно избавиться без каких-либо потерь для дальнейшего исследования.

```python
# Рассмотрим пропуски в столбце name
df[df['name'].isna()].head()
```

```python
# Удалим пропущенные значения в столбце name и проверим результат
df.dropna(subset=['name'], inplace=True)

print('Количество пропущенных значений в столбце name:', df['name'].isna().sum())
print('Количество пропущенных значений в столбце genre:', df['genre'].isna().sum())
```

Пропуски в столбце year_of_release составляют менее 2% и удаление этих данных не скажется отрицательно на проведении дальнейшего исследования.

```python
# Удалим пропущнные значения в столбце year_of_release 
df.dropna(subset=['year_of_release'], inplace=True)
```

```python
# Проверим, сколько пропусков осталось после замены
print('Количество пропусков в столбце year_of_release после замены:', df['year_of_release'].isna().sum())
```

Количество пропусков в столбце оценки критиков (critic_score) превышает 50%, в столбце оценки пользователей (user_score) - почти 55%, в столбце rating - 40%, что не позволяет нам их игнорировать, но и воостановить с особой точностью таже нет возможности, в виду нехватки информации. Посмотрим на уникальные значения каждого из параметров и заменим пропуски на условное значение, которое сможем в дальнейшем отнести к выбросам.

```python
# Посмотрим все уникальные значения столбца user_score
df['critic_score'].unique()
```

```python
# Посмотрим все уникальные значения столбца rating
df['rating'].unique()
```

Пропуски в рейтинге от организации ESRB, оценке пользователей и критиков заменим на условное значение, равное -1.

```python
# Заменим пропуски в столбце critic_score условным значением 
df['critic_score'] = df['critic_score'].fillna(-1)

# Заменим пропуски в столбце user_score условным значением 
df['user_score'] = df['user_score'].fillna(-1)

# Заменим пропуски в столбце rating условным значением 
df['rating'] = df['rating'].fillna(-1)

# Проверим количество пропусков  после замены
print('Количество пропусков в столбце critic_score после замены:', df['critic_score'].isna().sum())
print('Количество пропусков в столбце user_score после замены:', df['user_score'].isna().sum())
print('Количество пропусков в столбце rating после замены:', df['rating'].isna().sum())
```

### Часть 3. Исследовательский анализ данных <a id='3.'></a>

#### 3.1. Количество выпущенных игр в разные года <a id='3.1.'></a>

Посмотрим, сколько игр выпускалось в разные годы и постараемся определить нужны ли нам данные за все периоды.

```python
# Создадим сводную таблицу с указанием количества выпущенных игр по каждому году
games_per_year = df.pivot_table(
    index='year_of_release', 
    values='name', 
    aggfunc='count')

# Заменим название столбца name на count
games_per_year.columns = ['count']

# Проверим результат после замены
games_per_year.reset_index().transpose()
```

```python
# Построим столбчатую диаграмму, отражающую количество выпущенных игр по каждому году
plt.figure(figsize=(18,6)) 
sns.barplot(x=games_per_year.index, y="count", data=games_per_year)
plt.title("Количество выпущенных игр по каждому году")
plt.xlabel("Год выпуска")
plt.ylabel("Продано экземпляров игр, млн.")
plt.show()
```

С появлением в 1980-е годы персональных компьютеров начало формироваться направление разработки игр, но количество выпущенных игр не было столь велико, т.к. основой игровой индустрии все же продолжали оставаться аркадные автоматы и специализированные игровые приставки.

К увеличению доли компьютерных игр в 1991-1993 годах привело развитие аппаратных 3D технологии визуализации и появление основных игровых жанров. Так, например, в этот период была создана первая в мире игра в жанре «трёхмерный боевик» (3D-Action), первая стратегия в реальном времени (RTS), первая игра в жанре ужасов (Survival Horror).

Скачок роста количества выпускаемых компьютерных игр в 1994 году, связан с распространением операционных систем Microsoft и Mac OS. Также, компания «Sony» в этот год выпускает свою первую и очень удачную игровую консоль «PlayStation», которая обходит на рынке известного производителя игровых приставок Nintendo. В Ванкувере появляется первое учебное заведение, готовящее специалистов по разработке видео игр - DigiPen Institute of Technology и т.д.

Дальнейшее развитие игровой индустрии связано с широким распространением интернета, зарождением киберспорта, появлением мобильных устройств с достаточной производительностью для реализации игр и прочее.

```python
# Избавимся от данных за период с 1980 - по 1993 включительно
df = df.query('year_of_release >= 1994')
df.sample(5)
```

Исходя из истории развития индустрии компьютерных игр, для нас в исследовании не столь важны будут данные периода 1980 - 1993 годов включительно, что позволяет нам сразу их отбросить.

####  3.2. Динамика продаж по платформам в разные года <a id='3.2.'></a>

Рассмотрим суммарные продажи игр по консолям более подробно, найдем среди них лидеров и постараемся понять средний срок их жизни.

```python
# Найдем платформы с наибольшими суммарными продажами
max_total_sales = df.pivot_table(
    index='platform', 
    values='total_sales', 
    aggfunc='sum').sort_values(by='total_sales', ascending=False)

# Выведем данные на экран
max_total_sales.reset_index().transpose()
```

```python
# Построим столбчатую диаграмму, отражающую суммарные продажи по платформам
plt.figure(figsize=(18,6)) 
sns.barplot(x=max_total_sales.index, y=max_total_sales['total_sales'])
plt.title("Суммарные продажи по платформам")
plt.xlabel("Платформа")
plt.ylabel("Продано экземпляров игр, млн.")
plt.show()
```

Изучив более подробно суммарные объемы продаж игр, можно выделить несколько безусловных лидеров по данному показателю: PS2, X360, PS3, Wii, DS и PS.

```python
# Выведем на экран Топ-10 платформ с наибольшими продажами
top_10_platforms = max_total_sales.head(10).reset_index().transpose()
top_10_platforms
```

Если же рассматривать результаты по количеству проданных экземпляров в рамках рейтинга "Топ-10", то к первоначальным лидерам стоит добавить еще четырех: GBA, PS4, PSP	и 3DS.

```python
# Создадим сводную таблицу распределения продаж по годам на основе Топ-10 платформ
total_sales_of_top_10 = df.query('platform == ["PS2", "X360", "PS3", "WII", "DS", "PS", "GBA", "PS4", "PSP", "3DS"]').pivot_table(
    index='year_of_release', 
    columns='platform',
    values='total_sales', 
    aggfunc='sum')

# Выведем полученный результат на экран
total_sales_of_top_10.reset_index()
```

```python
# Построим график для большей наглядности
total_sales_of_top_10.plot(style='.-', kind='line', grid=True, fontsize=10, figsize=(16,6))
plt.title('Динамика продаж по платформам в разные года')
plt.xlabel('Год продажи')
plt.ylabel('Продано экземпляров игр, млн.')
plt.legend()
plt.show()
```

Рассмотрев более детально распределение продаж по годам можно сделать вывод, что в среднем "срок жизни" платформы около 10 лет, за это время они устаревают и на смену им приходят более современные прототипы. Таким образом, в качестве актуального периода для проведения данного исследования не обязательно брать все года, а можно ограничиться последними десятью.

#### 4.1. Топ-5 платформ по продажам для каждого региона <a id='4.1.'></a>

Найдем наиболее популярные платформы по продажам в Северной Америке, Европе и Японии, посчитав по каждому из них суммарные продажи и приходящиеся на них доли.

```python
# Напишем цикл для расчета суммарных продаж по каждому региону и долей, приходящихся на них
for column in ['na_sales', 'eu_sales', 'jp_sales']:
    location = actual_df.pivot_table(
        index='platform', 
        values=column, 
        aggfunc='sum').sort_values(column, ascending=False)
    
    location['sales_share'] = round(location[column] / location[column].sum(), 2)
   
    display(location.nlargest(5, column))
```

#### 4.3. Влияние рейтинга ESRB на продажи в отдельном регионе <a id='4.3.'></a>

```python
# Напишем цикл для расчета суммарных продаж по каждому региону и долей, приходящихся на них
for column in ['na_sales', 'eu_sales', 'jp_sales']:
    rating = actual_df.query('rating != -1').pivot_table(
        index='rating', 
        values=column, 
        aggfunc='sum').sort_values(column, ascending=False)
    
    rating['rating_share'] = round(rating[column] / rating[column].sum(), 2)
   
    display(rating.nlargest(5, column))
```

### Шаг 5. Проверка гипотез <a id='5.'></a>

Для проверки двух гипотез будем использовать t-критерий Стьюдента, т.к. выборки независимы между собой.

#### 5.1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые <a id='5.1.'></a>

H0: Средние пользовательские рейтинги платформ Xbox One и PC не различается.  
H1: Средние пользовательские рейтинги платформ Xbox One и PC различается.

```python
# Зададим переменные для Xbox One и PC
xone_rating = actual_df[(actual_df['platform'] == 'XONE') & (actual_df['user_score'] > 0)]['user_score']
pc_rating = actual_df[(actual_df['platform'] == 'PC') & (actual_df['user_score'] > 0)]['user_score']
```

```python
# Посчитаем дисперсию двух выборок
xone_variance = np.var(xone_rating)
print('Дисперсия выборки для платформы Xbox One:', xone_variance) 

pc_variance = np.var(pc_rating)
print('Дисперсия выборки для PC:', pc_variance) 

# Найдем отношение дисперсий двух выборок
variance = pc_variance/xone_variance
print('Отношение дисперсий:',variance)
```

Сравнивая дисперсии двух выборок мы видим, что они различаются более, чем на 5%. Отличие значимое, выбираем параметр equal_var = False.

```python
alpha = 0.05 # Зададим критический уровень статистической значимости
results = st.ttest_ind(xone_rating, pc_rating, equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
# Найдем средний пользовательский рейтинг по платформе Xbox One и РС
print('Средний пользовательский рейтинг по платформе Xbox One:', xone_rating.mean())
print('Средний пользовательский рейтинг по РС:', pc_rating.mean())
```

Заданный уровень значимости, равный 0.05, означает, что при проверки гипотезы допускается не более, чем 5% вероятность ошибки. 
В нашем случае мы не можем отвергнуть нулевую гипотезу в пользу альтернативной гипотезы, т.е. по результатам статистического теста вероятность различий в средних пользовательских рейтингах платформы Xbox One и PC не подтвердилась.

#### 5.2. Cредние пользовательские рейтинги жанров Action и Sports разные <a id='5.2.'></a>

H0: Средние пользовательские рейтинги жанров Action и Sports не различаются.  
H1: Средние пользовательские рейтинги жанров Action и Sports различаются.

```python
# Зададим переменую для жанров Action и Sports
action_genre = actual_df[(actual_df['genre'] == 'action') & (actual_df['user_score'] > 0)]['user_score']
sports_genre = actual_df[(actual_df['genre'] == 'sports') & (actual_df['user_score'] > 0)]['user_score']
```

```python
# Посчитаем дисперсию двух выборок
action_variance = np.var(action_genre)
print('Дисперсия выборки для жанра Action:', action_variance) 

sports_variance = np.var(sports_genre)
print('Дисперсия выборки для жанра Sports:', sports_variance) 

# Найдем отношение дисперсий двух выборок
variance = sports_variance/action_variance
print('Отношение дисперсий:',variance)
```

Сравнивая дисперсии двух выборок мы видим, что они различаются более, чем на 5%. Отличие значимое, выбираем параметр equal_var = False.

```python
alpha = 0.05 # Зададим критический уровень статистической значимости
results = st.ttest_ind(action_genre, sports_genre, equal_var=False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Статистический тест показал, что средние пользовательские рейтинги жанров Action и Sports различаются.

```python
# Найдем средний пользовательский рейтинг по жанрам Action и Sports
print('Средний пользовательский рейтинг по жанру Action:', action_genre.mean())
print('Средний пользовательский рейтинг по жанру Sports:', sports_genre.mean())
```

### Вывод

Проведя статистический анализ мы выяснили, что средние пользовательские рейтинги платформ Xbox One и PC одинаковы и средние пользовательские рейтинги для жанров Action и Sports различаются.
