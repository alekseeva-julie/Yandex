### Часть 1. Приоритизация гипотез <a id='1.'></a> 

На данном этапе постараемся решить задачу над какой из общего списка подготовленных гипотез для увеличения выручки мы будем работать в первую очередь. Для этого применим две наиболее популярные методологии приоритизации ICE и RICE и выясним, изменится ли как-то приоритизация гипотез при применении RICE вместо ICE.

Начнем уторядочивание с изучения исходного файла с гипотезами и оценками параметров ICE и RICE.

```python
# Импортируем необходимые библиотеки
import pandas as pd
import datetime as dt
import scipy.stats as stats
from scipy import stats as st
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Прочитаем данные hypothesis методом read_csv и посмотрим на исхдный датафейм
hypothesis = pd.read_csv('/datasets/hypothesis.csv')

# Приведем названия столбцов к нижнему регистру 
hypothesis.columns = hypothesis.columns.str.lower()

# Выведем результат
hypothesis
```

```python
# Изменим ширину колонки hypothesis методом max_colwidth
pd.set_option('display.max_colwidth', -1)
```

#### 1.1. Применение фреймворка ICE для приоритизации гипотез  <a id='1.1.'></a> 

Применим метод определения приоритетов для гипотез ICE и добавим полученные значения в новый столбец к существующей таблице.  
Будем использовать следующую формулу для расчета ICE:

$$ICE = \frac{Impact * Confidence}{Efforts}$$

```python
# Посчитаем ICE для гипотез и добавим полученные значения в новый столбец датафрейма
hypothesis['ICE'] = (hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']).round()

# Выведим столбцы 'hypothesis' и 'ICE', отсортированные по убыванию приоритета
hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
```

Оценка приоритетов в рамках метода ICE говорит о том, что наиболее перспективными гипотезами являются: 8, 0 и 7.  
Справедливым будет отметить, что по 10-балльной шкале:
* степень влияния данных идей на пользователей, их опыт и удовлетворение стремится или приходится на самые высокие показатели - 9, 10 и, соответственно, 7;  
* принимают наивысшие оценки в уверенности, что эти изменения повлияют на пользователей именно так;
* имеют средние показатели затрат на проверку гипотез, равные 5 и 6.

#### 1.2. Примениние фреймворка RICE для приоритизации гипотез <a id='1.2.'></a> 

Рассмотрим второй из самых популярных методов определения приоритетов - RICE и добавим полученные значения в новый столбец к существующей таблице.  
Для расчета RICE воспользуемся следующей формулой:  

$$RICE = \frac{Rearch * Impact * Confidence}{Efforts}$$

```python
# Посчитаем RICE для гипотез и добавим полученные значения в новый столбец датафрейма
hypothesis['RICE'] = hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']

# Выведим столбцы 'hypothesis' и 'RICE', отсортированные по убыванию приоритета
hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
```

По итогу оценки приоритетов методом RICE можно назвать гипотезы 7, 2, 0 и 6 наиболее перпективными.  
Отметим, что при использовании данного метода, на первом месте наблюдаем уже гипотезу 7. Такой результат мы получили, потому что ее параметр "охват пользователей" равен макисмальному значению по 10-балльной шкале, в то время как у других гипотез он меньше.

#### 1.3. Различия в приоритизации гипоз при применении метода RICE вместо ICE <a id='1.3.'></a> 

Применив две наиболее популярные методолгии приоритизации, ответим на вопрос, как изменились полученные показатели при применении RICE вместо ICE.

```python
# Выведем датафрейм hypothesis с показателями ICE и RICE
hypothesis
```

Несмотря на похожие с ICE показатели, в RICE еще применяют показатель Reach — количество пользователей, которые увидят изменения в продукте и на которых повлияет гипотеза. Так, например:
* занимающая лидирующую позицию в ICE, идея "Запустить акцию, дающую скидку на товар в день рождения", имея очень низкий показатель охвата, равный 1 по 10-балльной шкале, в RICE опускается на 4 позицию;
* идея "Добавить страницу отзывов клиентов о магазине, что позволит увеличить количество заказов" с третьего места ICE, с максимально возможным показателем охвата - 10/10, стала номером один в RICE.

Таким образом, несмотря на то, что данные фреймворки похожи по названиям, они все же отличаются. Кроме силы гипотезы, не будет лишним учитывать и то, как много пользователей она затронет.

$$ICE = \frac{Impact * Confidence}{Efforts}$$

$$RICE = \frac{Rearch * Impact * Confidence}{Efforts}$$

### Часть 2. Анализ A/B-теста <a id='2.'></a> 

Начнем анализ результатов А/В-теста с изучения данных из файлов с заказами - orders и пользователях - visitors.

```python
# Прочитаем данные orders методом read_csv и посмотрим на исхдный датафейм
orders = pd.read_csv('/datasets/orders.csv')

# Выведем первые 5 строк датафрейма
orders.head(5)
```

```python
# Приведем значения столбца 'date' к формату даты
orders['date'] = orders['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

# Проверим полученный результат
# orders.dtypes
```

```python
# Прочитаем данные visitors методом read_csv и посмотрим на исхдный датафейм
visitors = pd.read_csv('/datasets/visitors.csv')

# Выведем первые 5 строк датафрейма
visitors.head(5)
```

```python
# Приведем значения столбца 'date' к формату даты
visitors['date'] = visitors['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))

# Проверим полученный результат
# visitors.dtypes
```

Изучив файлы с данными о заказах и пользователях, а также приведя их к нужным типам там, где это было необходимо, далее приступим непосредственно к анализу A/B-теста.

#### 2.1. График кумулятивной выручки по группам  <a id='2.1.'></a> 

Для построения графика кумулятивной выручки по группам: создадим массив уникальных пар значений дат и групп теста, соберём агрегированные кумулятивные по дням данные о заказах и данные о посетителях.

```python
# Cоздадим массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()

# Соберём агрегированные кумулятивные по дням данные о заказах
ordersAggregated = datesGroups.apply(
    lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg({
        'date' : 'max',
        'group' : 'max',
        'transactionId' : pd.Series.nunique,
        'visitorId' : pd.Series.nunique,
        'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])

# Соберем агрегированные кумулятивные по дням данные о посетителях интернет-магазина
visitorsAggregated = datesGroups.apply(
    lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg({
        'date' : 'max',
        'group' : 'max',
        'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])

# Объединим кумулятивные данные в одну таблицу 
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])

# Переименуем названия столбцов
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

# Выведем 5 первых строк получившейся таблицы
cumulativeData.head(5) 
```

Построим графики кумулятивной выручки по дням и группам A/B-тестирования.

```python
# Объявим переменную cumulativeRevenueA, в которой сохраним данные о датах, выручке и числе заказов в группе A
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]

# Объявим переменную cumulativeRevenueB, в которой сохраним данные о датах, выручке и числе заказов в группе B
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

# Построим график выручки группы А и группы B
plt.figure(figsize=(16, 5))
plt.title('График кумулятивной выручки по группам')
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='A', color="red")
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='B')
plt.ylabel("Выручка")
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

Кумулятивный график выручки по группам говорит о том, что выручка увеличивается в течение всего теста - это хороший знак. Однако, в период теста с 18 - по 19 августа сегмент В вырвался вперед и продолжал лидировать весь остаток теста. Это может говорить о всплесках числа заказов, либо о появлении очень дорогих заказов в выборке. В дальнейшем потребуется обратить внимание на выбросы и проанализировать их.

#### 2.2. График кумулятивного среднего чека по группам <a id='2.2.'></a> 

Построим график среднего чека по группам. Для этого разделим кумулятивную выручку на кумулятивное число заказов.

```python
# Построим график среднего чека группы А и группы B
plt.figure(figsize=(16, 5))
plt.title('График кумулятивного среднего чека по группам')
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue']/cumulativeRevenueA['orders'], label='A', color="red")
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue']/cumulativeRevenueB['orders'], label='B')
plt.ylabel("Средний чек")
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

Кумулятивные значения среднего чека по группам характеризуются сильными колебаниями в первой половине теста. Однако, начиная с 18 августа и до конца теста средний чек для группы А стабилизировался примерно на одном уровне. Показатель же среднего чека группы В с 18 августа характеризуется резким всплеском и медленным спадом, что еще раз подтверждает версию попадания крупного заказа в данный сегмент. В целом, чтобы найти причину такого искажения результатов, потребуется более детально рассмотреть выбросы исходных данных.

#### 2.3. График относительного изменения кумулятивного среднего чека группы B к группе A <a id='2.3.'></a> 

Объединив таблицы cumulativeRevenueA и cumulativeRevenueB, построим график относительного различия для среднего чека группы B к группе A.

```python
# Объединим данные в одну таблицу
mergedCumulativeRevenue = cumulativeRevenueA.merge(cumulativeRevenueB, left_on='date', right_on='date', how='left', suffixes=['A', 'B'])

# Построим график отношения средних чеков
plt.figure(figsize=(16, 5))
plt.title('График относительного различия для среднего чека')
plt.plot(mergedCumulativeRevenue['date'], (mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB'])/(mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA'])-1, color="maroon")
plt.axhline(y=0, color='black', linestyle='--')
plt.ylabel("Относительное различие")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

В нескольких точках график различия между сегментами резко скачет. Так, мы можем наблюдать следующее:
* в первые несколько дней теста средний чек группы В к группе A показывает стремительный рост;
* спустя неделю мы уже наблюдаем падение показателей среднего чека группы В, которые становятся ниже показателей группы А на более чем 10%;  
* и, буквально через 5 дней, мы можем снова зафиксировать стремительное повышение показателей группы В.  

Данные резкие изменения указывают нам на наличие крупных заказов и выбросов.

#### 2.4. График кумулятивной конверсии по группам <a id='2.4.'></a> 

Посчитаем кумулятивную конверсию по дням по группам и построим график.

```python
# Посчитаем кумулятивную конверсию
cumulativeData['conversion'] = cumulativeData['orders']/cumulativeData['visitors']

# Сделаем срез по данным по группе A
cumulativeDataA = cumulativeData[cumulativeData['group']=='A']

# Сделаем срез по данным по группе B
cumulativeDataB = cumulativeData[cumulativeData['group']=='B']

# Построим график кумулятивной конверсии по дням по группам
plt.figure(figsize=(16, 5))
plt.title('График кумулятивной конверсии по дням по группам')
plt.plot(cumulativeDataA['date'], cumulativeDataA['conversion'], label='A', color="red")
plt.plot(cumulativeDataB['date'], cumulativeDataB['conversion'], label='B')
plt.grid()
plt.legend()
plt.ylabel("Конверсия")
plt.xticks(rotation=45)
plt.show()
```

В начале теста группа А имела большую конверсию, но спустя 5 дней конверсия сегмента В начала расти и зафиксировлась примерно на одном значении - от 3,3% - до 3,6%, в то время как конверсия группы А просела и также зафиксировалась на значениях - от 2,9% - до 3,1%.

#### 2.5. График относительного изменения кумулятивной конверсии группы B к группе A <a id='2.5.'></a> 

Построим график относительно различия кумулятивной конверсии группы B к группе A.

```python
# Объединим данные в одну таблицу
mergedCumulativeConversions = cumulativeDataA[['date','conversion']]\
    .merge(cumulativeDataB[['date','conversion']], left_on='date', right_on='date', how='left', suffixes=['A', 'B'])

# Построим график различия кумулятивной конверсии
plt.figure(figsize=(16, 5))
plt.title('График относительного изменения кумулятивной конверсии')
plt.plot(mergedCumulativeConversions['date'], mergedCumulativeConversions['conversionB']/mergedCumulativeConversions['conversionA']-1, label="Относительный прирост конверсии группы B относительно группы A", color="maroon")
plt.ylabel("Относительное различие")
plt.axhline(y=0, color='black', linestyle='--')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

В начале теста группа В значительно проигрывала по показателям конверсии группе А, всего один лишь раз показав прирост на 5% и снова потеряв данный результат. Однако, спустя 5 дней с начала теста конверсия группы В начинает постепенно расти и уже не опускается ниже прироста в 10%, а иногда даже достигает и более 20%. В целом же уже сейчас делать какие-либо выводы по тесту нельзя. Для более объективных выводов и результатов потребуется анализ аномалий.

#### 2.6. Точечный график количества заказов по пользователям <a id='2.6.'></a> 

В данном шаге приступим к работе с крайними значениями и погрешностями в исходных данных. Иными словами найдем аномалии, которые искажают результаты нашего А/В-теста.  
Анализ выбросов начнем с построения графика распределения числа заказов по пользователям.

```python
# Найдем число заказов по пользователям
ordersByUsers = orders.drop(['group', 'revenue', 'date'], axis=1).groupby('visitorId', as_index=False)\
            .agg({'transactionId' : pd.Series.nunique})
ordersByUsers.columns = ['userid','orders']

# Выведем первые 10 строк таблицы
ordersByUsers.sort_values(by='orders',ascending=False).head(10).reset_index()
```

```python
# Сгенерируем номера наблюдений из чисел от 0 до количества наблюдений в ordersByUsers
x_values = pd.Series(range(0,len(ordersByUsers)))

# Построим точечный график количества заказов по пользователям
plt.figure(figsize=(16, 5))
plt.title('Диаграмма распределения числа заказов по пользователям')
plt.scatter(x_values, ordersByUsers['orders'], color="olive") 
plt.ylabel("Количество заказов")
plt.show()
```

Более редкими значениями, отражающими количество покупок, совершенных одним пользователем, является - 2 и 3 заказа.

#### 2.10. Статистическая значимость различий в конверсии между группами по «сырым» данным <a id='2.10.'></a> 

В данном шаге посчитаем статистическую значимость различий в конверсии между группами по «сырым» данным — без удаления аномальных пользователей.  
Для этого сделаем срез по количеству пользователей в указанную дату для групп А и В; зададим переменные, в которых сохраним количество пользователей до выбранной даты по сегментам; создадим сводные таблицы с указанием количества заказов и суммарной выручкой в выбранную дату в группе А и В и сводные таблицы с суммарным числом заказов до выбранной даты включительно в каждой группе. Далее объединим все полученные результаты в одну таблицу.

```python
# Сделаем срез данных по дате и количеству пользователей в указанную дату по группе А
visitorsADaily = visitors[visitors['group']=='A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

# Сделаем срез данных по дате и количеству пользователей в указанную дату по группе В
visitorsBDaily = visitors[visitors['group']=='B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

# Объявим переменную visitorsACummulative, в которой сохраним данные о количестве посещений в день в группе А
visitorsACummulative = visitorsADaily.apply(lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']]\
                                            .agg({'date' : 'max', 'visitorsPerDateA' : 'sum'}), axis=1)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

# Объявим переменную visitorsACummulative, в которой сохраним данные о количестве посещений в день в группе В
visitorsBCummulative = visitorsBDaily.apply(lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']]\
                                            .agg({'date' : 'max', 'visitorsPerDateB' : 'sum'}), axis=1)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']

# Объявим ordersADaily с количеством заказов и суммарной выручкой в выбранную дату по группе А
ordersADaily = orders[orders['group']=='A'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

# Объявим ordersBDaily с количеством заказов и суммарной выручкой в выбранную дату по группе В
ordersBDaily = orders[orders['group']=='B'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

# Объявим ordersACummulative суммарным числом заказов до выбранной даты включительно в группе A
ordersACummulative = ordersADaily.apply(
    lambda x: \
        ordersADaily[ordersADaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateA' : 'sum',
                'revenuePerDateA' : 'sum'}), axis=1).sort_values(by=['date'])
ordersACummulative.columns = ['date', 'ordersCummulativeA', 'revenueCummulativeA']

# Объявим ordersBCummulative суммарным числом заказов до выбранной даты включительно в группе В
ordersBCummulative = ordersBDaily.apply(
    lambda x: \
        ordersBDaily[ordersBDaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateB' : 'sum',
                'revenuePerDateB' : 'sum'}), axis=1).sort_values(by=['date'])
ordersBCummulative.columns = ['date', 'ordersCummulativeB', 'revenueCummulativeB']

# Объединим полученные сводные таблицы в одну и назовем ее data
data = ordersADaily.merge(ordersBDaily, left_on='date', right_on='date', how='left')\
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')\
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')

# Выведем первые 5 строк и проверим поученный результат
data.head(5)
```

Посчитаем статистическую значимость различия в конверсии между группами, применим критерий Манна-Уитни.

```python
# Объявим переменную ordersByUsersA, где укажем число совершённых заказов для пользователей группы А, которые заказывали хотя бы 1 раз
ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False)\
                .agg({'transactionId' : pd.Series.nunique})
ordersByUsersA.columns = ['visitorId', 'orders']

# Объявим переменную ordersByUsersB, где укажем число совершённых заказов для пользователей группы В, которые заказывали хотя бы 1 раз
ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False)\
                .agg({'transactionId' : pd.Series.nunique})
ordersByUsersB.columns = ['visitorId', 'orders']

# Объявим переменную sampleA, в которой пользователям из группы А будет соответствовать количество заказов
sampleA = pd.concat([ordersByUsersA['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateA']\
                .sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)

# Объявим переменную sampleB, в которой пользователям из группы В будет соответствовать количество заказов
sampleB = pd.concat([ordersByUsersB['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateB']\
                .sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)
```

Сформулируем нулевую и альтернативную гипотезы различий в конверсии между группами по «сырым» данным:

H0: Между конверсиями групп А и В по «сырым» данным нет различий.  
H1: Между конверсиями групп А и В по «сырым» данным есть различия.

```python
# Зададим критический уровень статистической значимости
alpha = .05

# Найдём относительные различия в конверсии между группами по "сырым" данным
results = st.mannwhitneyu(sampleA, sampleB)

print('P-value конверсии между группами:', "{0:.5f}".format(results.pvalue))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу: вывод о различии сделать нельзя")

print('Относительное различие конверсии группы B по отношению к группе A:',"{0:.3f}".format((data['ordersPerDateB']\
            .sum()/data['visitorsPerDateB']\
            .sum())/(data['ordersPerDateA']\
            .sum()/data['visitorsPerDateA'].sum())-1)) 
```

Следует отметить, что полученное значение p-value намного меньше 0.05, что позволяет нам отвергнуть нулевую гипотезу. Анализ "сырых" данных свидетельствует о том, что в конверсии между группами есть статистически значимые различия. Относительный прирост конверсии группы В к конверсии группы А равен 13,8%.

#### 2.11. Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным <a id='2.11.'></a> 

Для расчета статистической значимости различий в среднем чеке между группами А и В, передадим критерию Манна-Уитни данные о выручке с заказов.  
Также, сформулируем нулевую и альтернативную гипотезы различий в среднем чеке между группами по «сырым» данным:

H0: Между средними чеками групп А и В по «сырым» данным нет различий.  
H1: Между средними чеками групп А и В по «сырым» данным есть различия.

```python
# Зададим критический уровень статистической значимости
alpha = .05

# Найдём относительные различия в среднем чеке между группами по "сырым" данным
results = st.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])

print('P-value среднего чека между группами:', "{0:.3f}".format(results.pvalue))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу: вывод о различии сделать нельзя")

print('Относительное различие среднего чека группы B по отношению к группе A:', "{0:.3f}".format(
    orders[orders['group']=='B']['revenue'].mean()/
    orders[orders['group']=='A']['revenue'].mean()-1)) 
```

P-value среднего чека между сегментами значительно больше 0.05. Следовательно, у нас нет причин отвергать нулевую гипотезу и считать, что в среднем чеке есть различия. Относительное различие среднего чека группы В по отношению к среднему чеку группы А составляет - 25,9%, что довольно много. Однако, не стоит забывать, что данный тест проводился на данных с аномальными значениями, в виду чего, пока рано говорить о его успешности. 

#### 2.12. Статистическая значимость различий в конверсии между группами по «очищенным» данным <a id='2.12.'></a> 

Перед тем, как приступить к анализу А/В-теста по "очищенным" данным, узнаем, сколько всего у нас пользователей, которых мы можем отнести к категории аномальных.  
Так, за аномальных пользователей мы примем тех, кто совершил 3 и более заказа или сделал заказ на сумму более 28 000 рублей.

```python
# Создадим таблицу usersWithManyOrders с пользователями, у которых более 2 покупок
usersWithManyOrders = pd.concat([ordersByUsersA[ordersByUsersA['orders'] > 2]['visitorId']\
                                 , ordersByUsersB[ordersByUsersB['orders'] > 2]['visitorId']], axis = 0)

# Создадим таблицу usersWithExpensiveOrders с пользователями, сделавшими заказ на сумму более 28000 рублей
usersWithExpensiveOrders = orders[orders['revenue'] > 28000]['visitorId']

# Объединим таблиы
abnormalUsers = pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis = 0).drop_duplicates().sort_values()

# Найдем количество аномальных пользователей
print('Количество аномальных пользователей:', len(abnormalUsers), 'человека') 
```

Узнаем, как их действия повлияли на результаты теста. Посчитаем статистическую значимость различий в конверсии между группами теста по "очищенным" данным.  
Сначала подготовим выборки количества заказов по пользователям по группам теста, а затем к полученным выборкам применим статистический критерий Манна-Уитни. 

```python
# Объявим переменную sampleAFiltered, в которых сохраним "очищенные" данные о заказах в группе А, исключая аномальных пользователей
sampleAFiltered = pd.concat([ordersByUsersA[np.logical_not(ordersByUsersA['visitorId']\
                .isin(abnormalUsers))]['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateA']\
                .sum() - len(ordersByUsersA['orders'])),name='orders')],axis=0)

# Объявим переменную sampleBFiltered, в которых сохраним "очищенные" данные о заказах в группе В, исключая аномальных пользователей
sampleBFiltered = pd.concat([ordersByUsersB[np.logical_not(ordersByUsersB['visitorId']\
                .isin(abnormalUsers))]['orders'],pd.Series(0, index=np.arange(data['visitorsPerDateB']\
                .sum() - len(ordersByUsersB['orders'])),name='orders')],axis=0) 
```

Сформулируем нулевую и альтернативную гипотезы различий в конверсии между группами по «очищенным» данным:

H0: Между конверсиями групп А и В по «очищенным» данным нет различий.  
H1: Между конверсиями групп А и В по «очищенным» данным есть различия. 

```python
# Зададим критический уровень статистической значимости
alpha = .05

# Найдём относительные различия в конверсии между группами по "очищенным" данным
results = st.mannwhitneyu(sampleAFiltered, sampleBFiltered)

print('P-value конверсии между очищенными группами:', "{0:.5f}".format(results.pvalue))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу: вывод о различии сделать нельзя")

print('Относительный прирост конверсии очищенной группы B по отношению к группе A:', "{0:.3f}".format(
    sampleBFiltered.mean()/sampleAFiltered.mean()-1)) 
```

Проведенный анализ "очищенных" данных говорит нам о том, что как и в случае с "сырыми" данными, мы достигли статистически значимых различий в конверсии между группами. Так, конверсия группы В составила 17,3%, что на 3,5% больше, чем конверсия между сегментами, где присутствуют аномальные пользователи.  
Из всего вышесказанного можно сделать вывод, что аномальные данные оказывают влияние на результаты А/В-теста. 

#### 2.13. Cтатистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным <a id='2.13.'></a>

Посчитаем статистическую значимость различия средних чеков после удаления аномальных пользователей и сформулируем нулевую и альтернативную гипотезы cтатистической значимости различий в среднем чеке между группами по «очищенным» данным:

H0: Между средними чеками групп А и В по «очищенным» данным нет различий.  
H1: Между средними чеками групп А и В по «очищенным» данным есть различия.

```python
# Зададим критический уровень статистической значимости
alpha = .05

# Найдём относительные различия в среднем чеке между группами по "очищенным" данным
results = st.mannwhitneyu(orders[np.logical_and(orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'],
        orders[np.logical_and(orders['group']=='B',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'])

print('P-value средних чеков между очищенными группами:', "{0:.3f}".format(results.pvalue))

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу: вывод о различии сделать нельзя")

print('Относительный прирост среднего чека очищенной группы B по отношению к группе A:', "{0:.3f}".format(
        orders[np.logical_and(orders['group']=='B',np.logical_not(orders['visitorId']\
        .isin(abnormalUsers)))]['revenue'].mean()/
        orders[np.logical_and(orders['group']=='A',
        np.logical_not(orders['visitorId'].isin(abnormalUsers)))]['revenue'].mean() - 1))
```

Сравнивая статистическую значимость различий среднего чека до удаления аномальных пользователей и после их удаления, можно сделать вывод, что результаты практически не изменились. Но стоит обратить внимание, что избавившись от выбросов, в наших данных стало меньше шума, благодаря чему мы смогли увидеть, что средний чек группы В стал меньше среднего чека группы А на 2%. Данный показать существенно отличается от значения, полученного на "сырых" данных, что сильно повлияет на итоговый результат и позволит дать объективную оценку по итогу проведенного анализа А/В-теста.

### Общий вывод

По результатам проведенного А/В-теста сделаем следующие выводы:
* график различия конверсии между сегментами сообщает, что результаты группы B намного лучше результатов группы A. Показатели сегмента В на протяжении длительного времени не опускаются ниже прироста в 10% и имеют значительные тенденции к улучшению;
* график различия среднего чека колеблется и он то как раз и позволил нам найти аномалии;
* избавление от шума в данных дало нам возможность увидеть, что средний чек в группе В меньше среднего чека группы А на 2%;
* есть статистически значимое различие по конверсии между группами как по данным, где присутствуют аномальные пользователи, так и после фильтрации аномалий. Так, конверсия группы В после очистки данных составила почти 18%.

Исходя из обнаруженных фактов, тест следует остановить и зафиксировать победу группы В над группой А.
